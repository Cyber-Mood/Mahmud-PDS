#Task 1

1. Create two large DataFrames (df1 and df2) with 1 million rows and 10 columns,
filled with random float values.
2. Perform the following operations:  
● Add df1 and df2  
● Multiply df1 and df2  
● Subtract df2 from df1  
3. Measure and compare the execution time and memory usage for each operation
using %timeit.

````python
#1
import pandas as pd
import numpy as np

df1 = pd.DataFrame(np.random.rand(1_000_000, 10), columns=[f"col_{i}" for i in range(10)])
df2 = pd.DataFrame(np.random.rand(1_000_000, 10), columns=[f"col_{i}" for i in range(10)])
#df1
#df2
````

````python
#2
# Addition
df_add = df1 + df2
#df_add

# Multiplication
df_mul = df1 * df2
#df_mul

# Subtraction
df_sub = df1 - df2
#df_sub
````

````python
#3
# Measure addition exec time and memory usage
%timeit df1 + df2
add_mem = df_add.memory_usage(deep=True).sum()

# Measure multiplication exec time and memory usage
%timeit df1 * df2
mul_mem = df_mul.memory_usage(deep=True).sum()

# Measure subtraction exec time and memory usage
%timeit df1 - df2
sub_mem = df_sub.memory_usage(deep=True).sum()

print(f"Addition memory usage: {add_mem} bytes") # for better readability bytes --> MB
print(f"Multiplication memory usage: {mul_mem} bytes")
print(f"Subtraction memory usage: {sub_mem} bytes")
````

#Task 6

You are given this dataset:  
```
data = pd.DataFrame({
'price': [250000, np.nan, 150000, 300000, 450000, np.nan, 150000],
'area': ['2000 sqft', '1800 sqft', '1500 sqft', np.nan, '2500 sqft', '2000 sqft', '1500 sqft'],
'bedrooms': [3, 2, 2, 4, np.nan, 3, 2],
'location': ['New York', 'Los Angeles', 'New York', 'New York', 'Chicago', 'Los Angeles', 'New York'],
'date': ['2022-01-10', '01/15/2022', '2022-01-20', '01-25-2022', '2022-02-01', '2022-02-05', '2022-02-10']
})
```
1. Identify and handle missing values using appropriate methods.
2. Identify and remove any duplicate entries.
3. Standardize columns like area (remove “sqft” and convert to numeric), and ensure
dates are consistent.

````python
#Given dataset

import pandas as pd
import numpy as np

data = pd.DataFrame({
    'price': [250000, np.nan, 150000, 300000, 450000, np.nan, 150000],
    'area': ['2000 sqft', '1800 sqft', '1500 sqft', np.nan, '2500 sqft', '2000 sqft', '1500 sqft'],
    'bedrooms': [3, 2, 2, 4, np.nan, 3, 2],
    'location': ['New York', 'Los Angeles', 'New York', 'New York', 'Chicago', 'Los Angeles', 'New York'],
    'date': ['2022-01-10', '01/15/2022', '2022-01-20', '01-25-2022', '2022-02-01', '2022-02-05', '2022-02-10']
})
print(data)
````

````python
#1
#Check for missing values
print("Missing values in each column:  ")
print(data.isna().sum())

#handling
data['price'] = data['price'].fillna(data['price'].median())
data['area'] = data['area'].fillna(data['area'].mode()[0])
data['bedrooms'] = data['bedrooms'].fillna(data['bedrooms'].mode()[0])

print("\nAfter handling missing values")
print(data)
````

````python
#2
#Check Duplicates
print(f"Number of duplicates: {data.duplicated().sum()}")

#remove duplicates
data = data.drop_duplicates()
print("\nAfter removing duplicates")
print(data)
````

````python
#3
# Area Standardizing
data['area'] = data['area'].astype(str)  # Convert all values to strings
data['area'] = pd.to_numeric(data['area'].str.replace(r'\D+', '', regex=True), errors='coerce')

# Date consistent
data['date'] = pd.to_datetime(data['date'], errors='coerce')

print("\nAfter Standardizing Area and Date:")
print(data)
````
